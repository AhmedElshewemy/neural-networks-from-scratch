```python
import numpy as np
import urllib.request

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/optdigits/optdigits.tra"
urllib.request.urlretrieve(url, "optdigits.tra")

def load_data(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    data = []
    for line in lines:
        data.append([int(x) for x in line.strip().split(',')])
    return np.array(data)

data = load_data("optdigits.tra")

X = data[:, :-1]
y = data[:, -1]
```

```python
def softmax(x):
  x=x-np.max(x,axis=1,keepdims=True)
  exp_x=np.exp(x)
  return exp_x/np.sum(exp_x,axis=1,keepdims=True)

def cross_entropy(y_true,y_pred):
  y_pred=np.clip(y_pred,1e-12,1 - 1e-12)
  return -np.mean(np.sum(np.log(y_pred)*y_true,axis=1))
def gradient_descent (X , y_true , y_pred , w , b , learning_rate ) :
    m = X . shape [0]

    dw = (1/ m ) * np . dot ( ( y_pred - y_true ).T , X )

    db = (1/ m ) * np . sum ( y_pred - y_true , axis =0)

    w = w - learning_rate * dw
    b = b - learning_rate * db
    return w , b

def cross_entropy_loss( y_true,y_pred):
      num_samples = len(y_true)
      return -np.sum(np.log(y_pred[range(num_samples), y_true])) / num_samples
def one_hot_encode(y):
      one_hot = np.zeros((len(y), num_classes))
      for i, val in enumerate(y):
          one_hot[i, val] = 1
      return one_hot
def predict(X, weights,b):
    scores = np.dot(X, weights.T)+b
    probs = softmax(scores)
    return np.argmax(probs, axis=1)
```
