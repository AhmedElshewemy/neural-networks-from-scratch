```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical

# Load dataset
data = np.loadtxt("https://archive.ics.uci.edu/ml/machine-learning-databases/optdigits/optdigits.tra", delimiter=",")

# Shuffle the data
np.random.shuffle(data)

# Split data into features and labels
X = data[:, :-1]
y = data[:, -1]

# Normalize the features
X /= 16

# One-hot encode the labels
y_one_hot = to_categorical(y)

# Split data into train and test sets
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y_one_hot[:split], y_one_hot[split:]

# Define the neural network architecture
model = Sequential([
    Dense(128, activation='sigmoid', input_shape=(X.shape[1],)),
    Dense(10, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2)

# Evaluate on test set
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Loss: {loss}, Test Accuracy: {accuracy}")
```
